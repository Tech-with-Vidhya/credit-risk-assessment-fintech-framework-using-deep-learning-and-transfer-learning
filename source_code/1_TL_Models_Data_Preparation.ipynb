## Data Preparation : Source Model (SM) and Target Model (TM) for Transfer Learning

#### Splitting the data into two sets of 50% each - one for the Source Model (SM) and the other for the Target Model (TM)

```python
# Source Model (SM) Data

sm_credit_risk_raw_data = credit_risk_data_all_shuffled.sample(frac=0.5, random_state=8, axis=0)
print("Total Data Instances: ", len(credit_risk_data_all_shuffled))
print("Source Model (SM) Data Instances: ", len(sm_credit_risk_raw_data))     
    
print("Execution Completed")
```

```python
# Resetting the index of the Source Model (SM) data

sm_credit_risk_raw_data.reset_index(drop=True, inplace=True)

print("Execution Completed")
```

```python
# Checking the size of the Source Model (SM) dataset

sm_credit_risk_raw_data.shape
```

```python
# Displying the last 5 data instances of the Source Model (SM) Data

sm_credit_risk_raw_data.tail()
```

```python
# Target Model (TM) Data

tm_credit_risk_raw_data = credit_risk_data_all_shuffled.drop(sm_credit_risk_raw_data.index)
print("Total Data Instances: ", len(credit_risk_data_all_shuffled))
print("Target Model (TM) Data Instances: ", len(tm_credit_risk_raw_data))

print("Execution Completed")
```

```python
# Resetting the index of the Target Model (TM) data

tm_credit_risk_raw_data.reset_index(drop=True, inplace=True)

print("Execution Completed")
```
