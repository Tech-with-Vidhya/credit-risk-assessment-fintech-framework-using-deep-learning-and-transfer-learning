## Target Model 1 (TM1) - Accuracy and Error Performance Metrics

```python
# Adding a new column "Credit_Score_Deviation_Status" to track whether the deviation between the actuals and predicted Credit Scores

# If the deviation is 0; then mark new column "Score_Deviation_Status" as "Perfect".
# If the deviation is greater than 0; then mark new column "Score_Deviation_Status" as "Error".

tm1_reg_results['Credit_Score_Deviation_Status'] = np.where(tm1_reg_results['Credit_Score_Deviation'] == 0, "Perfect", "Error")

# Displaying the first 10 data instances of the data frame
tm1_reg_results.head(10)
```

### Primarily; as input features related to Non-FICO factor named "Lenders Internal" Factor is not in scope of the Actual FICO Scores (as per FICO's Official Scoring System) and this proposed custom credit scoring model includes "Lenders Internal" Factors; hence it is a sensible decision to have a trade-off in the deviation of the predicted scores from that of the actual scores in order to verify the number of predictions that are closer to the actuals and also decide on a potential optimal credit scoring model with an acceptable error/deviation.

### Secondly to add; the finalised model has a validation loss of about 50.7289 which again could potentially be due to the non-inclusion of NON-FICO Lender's Internal Factor as part of the actual FICO Scores.

### Hence by considering the above two kep points; we can choose a value of about <b><font color='purple'>75 (25% out of the minimum score of 300)</font></b> as the Minimum Acceptable Deviation in the Credit Score by considering the point that Non-FICO Lender's Internal Factors contribute to a custom percentage of about 25% (5% discounted from each of the 5 FICO Factors Weightages) of the overall 100% Credit Score; and FICO Factors contribute to the rest of the 75% of the overall Credit Score.

> #### 1. FICO Factor 1 - Payment History - <b><font color='blue'>(35%)</font></b> - After Discounting 5% <b><font color='green'>- (30%)</font></b>
> #### 2. FICO Factor 2 - Amount Owed - <b><font color='blue'>(30%)</font></b> - After Discounting 5% <b><font color='green'>- (25%)</font></b>
> #### 3. FICO Factor 3 - Length of Credit History - <b><font color='blue'>(15%)</font></b> - After Discounting 5% <b><font color='green'>- (10%)</font></b>
> #### 4. FICO Factor 4 - Credit Mix - <b><font color='blue'>(10%)</font></b> - After Discounting 5% <b><font color='green'>- (5%)</font></b>
> #### 5. FICO Factor 5 - New Credit - <b><font color='blue'>(10%)</font></b> - After Discounting 5% <b><font color='green'>- (5%)</font></b>
> #### 6. Non-FICO Factor 6 - Lender's Internal - <b><font color='blue'>(Custom)</font></b> <b><font color='green'>- (25%)</font></b>

### Hence adding a new column "Credit_Score_Deviation_Status" to track whether the deviation between the actuals and predicted Credit Scores are less than or equal to 75 or not. 

> ### If the deviation is less than or equal to 75; then mark new column "Credit_Score_Deviation_Status" as "Accept".
> ### If the deviation is greater than 75; then mark new column "Score_Deviation_Status" as "Error".
> ### If the deviation is 0; then mark new column "Score_Deviation_Status" as "Perfect".

```python
# Primarily as input features related to Non-FICO factor named "Lenders Internal" Factor is not in scope of the Actual FICO Scores (as per FICO's Official 
# Scoring System) and this proposed custom credit scoring model includes "Lenders Internal" Factors; hence it is a sensible decision to have a trade-off 
# in the deviation of the predicted scores from that of the actual scores in order to verify the number of predictions that are closer to the actuals and 
# also decide on a potential optimal credit scoring model with an acceptable error/deviation.

# Secondly to add; the finalised model has a validation loss of about 50.7289 which again could potentially be due to the non-inclusion of NON-FICO 
# Lender's Internal Factor as part of the actual FICO Scores.

# Hence by considering the above two kep points; we can choose a value of about 75 as the Minimum Acceptable Deviation in the Credit Score by considering 
# the point that Non-FICO Lender's Internal Factors contribute to a custom percentage of about 25% (5% discounted from each of the 5 FICO Factors Weightages) of the overall 100% Credit Score; 
# and FICO Factors contribute to the rest of the 75% of the overall Credit Score.

# Hence adding a new column "Credit_Score_Tradeoff_Deviation_Status" to track whether the deviation between the actuals and predicted Credit Scores are 
# less than or equal to 75 or not. 

# If the deviation is less than or equal to 75; then mark new column "Credit_Score_Deviation_Status" as "Accept".
# If the deviation is greater than 75; then mark new column "Score_Deviation_Status" as "Error".
# If the deviation is 0; then mark new column "Score_Deviation_Status" as "Perfect".


tm1_reg_results['Credit_Score_Tradeoff_Deviation_Status'] = np.where(tm1_reg_results['Credit_Score_Deviation'] <= 75 , "Trade-off", "Error")
tm1_reg_results['Credit_Score_Tradeoff_Deviation_Status'] = np.where(tm1_reg_results['Credit_Score_Deviation'] == 0, "Perfect", tm1_reg_results['Credit_Score_Tradeoff_Deviation_Status'])

# Displaying the first 10 data instances of the data frame
tm1_reg_results.head(10)
```

```python
# Extracting the Total Count of "Perfect" and "Error" Credit Score from the "Credit_Score_Deviation_Status" Column before applying Deviation Trade-off

tm1_deviation_status_count = tm1_reg_results['Credit_Score_Deviation_Status'].value_counts()
tm1_deviation_status_count
```

```python
# Extracting the Total Count of "Perfect", "Trade-off" and "Error" Credit Score from the "Credit_Score_Deviation_Status" Column after applying 
# Deviation Trade-off

tm1_deviation_status_tradeoff_count = tm1_reg_results['Credit_Score_Tradeoff_Deviation_Status'].value_counts()
tm1_deviation_status_tradeoff_count
```

```python
# Calculating the Percentage of Accuracy and Error without appying trade-of in terms of the deviations of the predicted scores from that of the actual scores.

tm1_reg_accuracy_percentage = tm1_deviation_status_count['Perfect'] / tm1_reg_results.shape[0] * 100
print("Target Model 1 - Accuracy in Percentage Before Applying Trade-off Deviation: ", tm1_reg_accuracy_percentage)

tm1_reg_error_percentage = tm1_deviation_status_count['Error'] / tm1_reg_results.shape[0] * 100
print("Target Model 1 - Error in Percentage Before Applying Trade-off Deviation: ", tm1_reg_error_percentage)
```

```python
# Calculating the Percentage of Accuracy and Error based on the Acceptable Custom Trade-off Deviation Value set as an accepted trade-off in terms of the 
# deviations of the predicted scores from that of the actual scores.

tm1_reg_tradeoff_accuracy_percentage = (tm1_deviation_status_tradeoff_count['Trade-off'] + tm1_deviation_status_tradeoff_count['Perfect']) / tm1_reg_results.shape[0] * 100
print("Target Model 1 - Accuracy in Percentage After Applying Trade-off Deviation: ", tm1_reg_tradeoff_accuracy_percentage)

tm1_reg_tradeoff_error_percentage = tm1_deviation_status_tradeoff_count['Error'] / tm1_reg_results.shape[0] * 100
print("Target Model 1 - Error in Percentage After Applying Trade-off Deviation: ", tm1_reg_tradeoff_error_percentage)
```

### As we can see from the above results; the selected finalised proposed Credit Risk Scoring Regression Target Model (TM1) have an accuracy of about 77.13% with the inclusion of the trade-off in the credit score deviation between the actuals and the predicted values by considering an impact/contribution of about 25% by the Non-FICO Lender's Internal Factors to the overall credit scores.
