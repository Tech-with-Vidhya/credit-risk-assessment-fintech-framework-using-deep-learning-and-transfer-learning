## Splitting the Source Model (SM) Data in to 70% Train Data, 24% Validation Data and 6% Test Data

```python
# Shuffling the dataframe randomly based on the data instances (rows)

source_model_data_shuffled = source_model_data.sample(frac=1, random_state=5, axis=0)

print("Execution Completed")
```

```python
# Source Model (SM) Training Data

source_model_train = source_model_data_shuffled.sample(frac=0.7, random_state=8, axis=0)
print("Total Data Instances: ", len(source_model_data_shuffled))
print("Source Model (SM) Training Data Instances: ", len(source_model_train))     
    
print("Execution Completed")
```

```python
# Source Model (SM) Remaining Data

source_model_remaining = source_model_data_shuffled.drop(source_model_train.index)
print("Total Data Instances: ", len(source_model_data_shuffled))
print("Source Model (SM) Remaining Data Instances: ", len(source_model_remaining))

print("Execution Completed")
```

```python
# Source Model (SM) Validation Data

source_model_validation = source_model_remaining.sample(frac=0.8, random_state=12, axis=0)
print("Total Remaining Data Instances: ", len(source_model_remaining))
print("Source Model (SM) Validation Data Instances: ", len(source_model_validation))     
    
print("Execution Completed")
```

```python
# Source Model (SM) Test Data

source_model_test = source_model_remaining.drop(source_model_validation.index)
print("Total Remaining Data Instances: ", len(source_model_remaining))
print("Source Model (SM) Test Data Instances: ", len(source_model_test))

print("Execution Completed")
```

```python
# Verifying the size of the Training, Validation and Test Data Splits of the Source Model (SM) Dataset

print("Source Model Total Data Size: ", source_model_data.shape)


print("\nSource Model Training Data Size: ", source_model_train.shape)
print("Source Model Validation Data Size: ", source_model_validation.shape)
print("Source Model Test Data Size: ", source_model_test.shape)
```

```python
# Reset the index of the Training Data of the Source Model (SM) Dataset

source_model_train.reset_index(drop=True, inplace=True)

print("Execution Completed")
```

```python
# Reset the index of the Validation Data of the Source Model (SM) Dataset

source_model_validation.reset_index(drop=True, inplace=True)

print("Execution Completed")
```

```python
# Reset the index of the Test Data of the Source Model (SM) Dataset

source_model_test.reset_index(drop=True, inplace=True)

print("Execution Completed")
```
