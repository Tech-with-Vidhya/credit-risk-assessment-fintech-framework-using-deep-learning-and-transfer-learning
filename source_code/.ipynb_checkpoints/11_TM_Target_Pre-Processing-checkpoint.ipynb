## 1. Creating and Pre-Processing the new Target (Dependent) Variable 1 - "credit_score"

### Creation of a new Target Variable Column named "credit_score" for the Target Model (TM)

### Creating a new column named "FICO Average Latest" by taking the average of the 2 attributes "last_fico_range_high" and "last_fico_range_low"

```python
# Initially updating all the values of the "fico_average_latest" column with the average/mean of the latest/last fico score low and high values

fico_average_latest = (target_model_raw_data['last_fico_range_low'] + target_model_raw_data['last_fico_range_high'])/2
target_model_raw_data['fico_average_latest'] = fico_average_latest

print("Execution Completed")
```

```python
# Displying the first 5 data instances of the Source Model (SM) Data

target_model_raw_data.head()
```

```python
# Conditional handling of zero values in "last_fico_range_low" column or "last_fico_range_high" column or in both the columns

#condition_1 = (target_model_raw_data['last_fico_range_low'] != 0.0) & (target_model_raw_data['last_fico_range_high'] == 0.0)
#condition_2 = (target_model_raw_data['last_fico_range_low'] == 0.0) & (target_model_raw_data['last_fico_range_high'] != 0.0)
#condition_3 = (target_model_raw_data['last_fico_range_low'] == 0.0) & (target_model_raw_data['last_fico_range_high'] == 0.0)
#condition_4 = (target_model_raw_data['last_fico_range_low'] != 0.0) & (target_model_raw_data['last_fico_range_high'] != 0.0)

# Set the value of "fico_average_latest" to the value in "last_fico_range_low"; if "last_fico_range_high" value is zero
target_model_raw_data['fico_average_latest'] = np.where((target_model_raw_data['last_fico_range_low'] != 0.0) & (target_model_raw_data['last_fico_range_high'] == 0.0), target_model_raw_data['last_fico_range_low'], target_model_raw_data['fico_average_latest'])

# Set the value of "fico_average_latest" to the value in "last_fico_range_high"; if "last_fico_range_low" value is zero
target_model_raw_data['fico_average_latest'] = np.where((target_model_raw_data['last_fico_range_low'] == 0.0) & (target_model_raw_data['last_fico_range_high'] != 0.0), target_model_raw_data['last_fico_range_high'], target_model_raw_data['fico_average_latest'])

# Last Observation Carried Forward (LOCF) Method
# Set the value of "fico_average_latest" to the mean value of "fico_range_low" and "fico_range_high"; if both "last_fico_range_low" and "last_fico_range_high" values are zero
target_model_raw_data['fico_average_latest'] = np.where((target_model_raw_data['last_fico_range_low'] == 0.0) & (target_model_raw_data['last_fico_range_high'] == 0.0), (target_model_raw_data['fico_range_low'] + target_model_raw_data['fico_range_high'])/2, target_model_raw_data['fico_average_latest'])

# Set the value of "fico_average_latest" to the mean value of "last_fico_range_low" and "last_fico_range_high"; if both "last_fico_range_low" and "last_fico_range_high" values are not equal to zero
target_model_raw_data['fico_average_latest'] = np.where((target_model_raw_data['last_fico_range_low'] != 0.0) & (target_model_raw_data['last_fico_range_high'] != 0.0), (target_model_raw_data['last_fico_range_low'] + target_model_raw_data['last_fico_range_high'])/2, target_model_raw_data['fico_average_latest'])


print("Execution Completed")
```

## 2. Creating and Pre-Processing the new Target (Dependent) Variable 2 - "credit_default"

### Creating a new target column named " credit_default" to predict the probability of loan default of an individual based on the existing column "loan_status"

> ### "credit_default" = 0 represents not a default and good
> ### "credit_default" = 1 represents default and bad

```python
# Checking the unique values of the "loan_status" column

target_model_raw_data['loan_status'].values
```

```python
# Creating a new Target Variable Column

target_model_raw_data['credit_default'] = np.where(target_model_raw_data.loc[:, 'loan_status'].isin(['Charged Off', 'Default', 'Late (31-120 days)',
                                                                       'Does not meet the credit policy. Status:Charged Off']), 1, 0)

print("Execution Completed")
```
