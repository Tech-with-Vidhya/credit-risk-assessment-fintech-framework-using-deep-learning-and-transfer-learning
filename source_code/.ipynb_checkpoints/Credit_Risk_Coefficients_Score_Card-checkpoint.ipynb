```python
# A Function to Print the Regression Co-efficients in a readable equation format

def reg_coefs_display(coefs, names = None, sort = False):
    if names == None:
        names = ["X%s" % x for x in range(len(coefs))]
    lst = zip(coefs, names)
    if sort:
        lst = sorted(lst,  key = lambda x:-np.abs(x[0]))
    return " + ".join("%s * %s" % (coef, name)
                                   for coef, name in lst)
```

```python
print ("Ridge model:", reg_coefs_display(ridge.coef_))
```
