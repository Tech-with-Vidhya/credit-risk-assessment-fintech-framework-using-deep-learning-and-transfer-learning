## Loading the Pre-Trained Target Model 1 (TM1) for Regression Prediction of "credit_score" as per Scenario-2

### Target Model 1 (TM1) as per Scenario-2 includes 
> ### 1 input layer with 74 neurons
> ### 1 hidden layer with 74 neurons and with "relu" activation function
> ### 1 output layer with a single neuron with "linear" activation function for regression prediction output of the "credit_score"

```python
from tensorflow.keras.models import Model
from tensorflow.keras.models import load_model

# Loading the Pre-Trained Target Model 1 (TM1) for Regression Prediction of "credit_score" as per Scenario-2

tm1_model_path = "/Users/vidhyalakshmiparthasarathy/.CMVolumes/Google-Drive-pbvidhya/QMUL_MSc_Project/credit_risk_scoring_project/models/target_model/tm_model_1_regression_tune_2_retrain_1_100000b_final/tm_model_1_regression_tune_2_retrain_1_100000b_final.h5"

target_model_1_pre_trained = load_model(tm1_model_path)

print("Target Model Loaded Successfully")
```

```python
# Verifying the Summary of the Model

target_model_1_pre_trained.summary()
```

## Target Model 1 (TM1) - Validation Data Predictions - Regression - "credit_score" as per Scenario-2

```python
# Predicting the Target "Credit Score" on the Validation Data

tm1_y_pred = target_model_1_pre_trained.predict(tm1_validation_X, batch_size=1)

print("Execution Completed")
```

```python
# Verifying the Count of Prediction Outcomes on the Validation Data

print("Total Output Actuals on the Target Model 1 Validation Data: ", tm1_validation_y.shape[0])
print("Total Output Predictions on the Target Model 1 Validation Data: ", len(tm1_y_pred))
```

```python
# Comparing the Actual and Prediction Results for the first 10 Test Validation Data Instances

print("Target Model 1 Validation Actuals")
print(tm1_validation_y[:10])

print("Target Model 1 Validation Predictions")
print(tm1_y_pred[:10])
```

## Target Model 1 (TM1) - Actuals versus Predictions Comparison as per Scenario-2

```python
# Converting credit score predictions numpy array into a dataframe

tm1_y_pred_df = pd.DataFrame(tm1_y_pred, columns=['Credit_Score_Prediction'])
tm1_y_pred_df.head()
```

```python
# Verifying the Size and the Missing Values (if any) Count of the Credit Default Actual (Ground-Truth) Dataframe and the Predictions Dataframe

print("Total Number of Observations in the Validation Data Actuals: ", tm1_validation_y.shape)
print("Total Number of Observations in the Validation Data Predictions: ", tm1_y_pred_df.shape)

print("\n")

print("Total Number of Missing Values in the Validation Data Actuals: ", tm1_validation_y.isna().sum())
print("Total Number of Missing Values in the Validation Data Predictions: ", tm1_y_pred_df.isna().sum())
```

```python
# Concatenating the Credit Score Actual (Ground-Truth) Dataframe and the Predictions Dataframe

frames = [tm1_validation_y, tm1_y_pred_df]

tm1_results = pd.concat(frames, axis=1)

# Renaming the "credit_score" column
tm1_results.rename(columns = {'credit_score':'Credit_Score_Actual'}, inplace = True)

# Displaying a sample of 10 random results
tm1_results.sample(10)
```

```python
# Checking the Data Types of the Columns in the Concatenated Results DataFrame

tm1_results.dtypes
```

```python
# Converting the float data type column "Credit_Score_Prediction" to integer data type

tm1_results = tm1_results.round(0).astype(int)

# Re-checking the data types of the results dataframe after converting to integer data type
print("Modified Data Types of the Results DataFrame: ", tm1_results.dtypes)
print("\n")

# Displaying a sample of 10 random results
tm1_results.sample(10)
```

## Target Model 1 (TM1) - Analysing the Predictions - Scenario-2

```python
# Adding a New Column to Track the Deviation Between the Actuals and Predicted Credit Scores

tm1_results['Credit_Score_Deviation'] = tm1_results['Credit_Score_Actual'].sub(tm1_results['Credit_Score_Prediction'], axis=0)

# Displaying the Results of the First 10 Data Instances
tm1_results[:10]
```

```python
# Transforming the Deviation Value to Absolute Value
tm1_results['Credit_Score_Deviation'] = tm1_results.Credit_Score_Deviation.abs()

# Displaying the results of first 10 data instances
tm1_results.head(10)
```

## Target Model 1 (TM1) - Accuracy and Error Performance Metrics - Scenario-2

```python
# Adding a new column "Credit_Score_Deviation_Status" to track whether the deviation between the actuals and predicted Credit Scores

# If the deviation is 0; then mark new column "Score_Deviation_Status" as "Perfect".
# If the deviation is greater than 0; then mark new column "Score_Deviation_Status" as "Error".

tm1_results['Credit_Score_Deviation_Status'] = np.where(tm1_results['Credit_Score_Deviation'] == 0, "No Deviation", "Deviation")

# Displaying the first 10 data instances of the data frame
tm1_results.head(10)
```

```python
# Extracting the Total Count of "No Deviation" and "Deviation" Credit Score from the "Credit_Score_Deviation_Status" Column before applying Deviation Trade-off

tm1_deviation_status_count = tm1_results['Credit_Score_Deviation_Status'].value_counts()
tm1_deviation_status_count
```

### Primarily; as input features related to Non-FICO factor named "Lenders Internal" Factor is not in scope of the Actual FICO Scores (as per FICO's Official Scoring System) and this proposed custom credit scoring model includes "Lenders Internal" Factors; hence it is logically not appropriate to expect the outcomes of the custom model's predictions to be equivalent to the fico score actuals; and the deviation between them is acceptable and justifiable.

### Secondly to add; the validation loss encountered during the model training step; could potentially be due to the non-inclusion of NON-FICO Lender's Internal Factor as part of the actual FICO Scores.
