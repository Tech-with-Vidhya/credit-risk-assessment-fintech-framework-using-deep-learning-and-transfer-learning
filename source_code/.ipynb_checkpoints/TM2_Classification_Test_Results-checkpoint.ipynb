## Loading the Optimised Pre-Trained Target Model 2 (TM2) for Classification Evaluation of "credit_default"

### Target Model 2 (TM2) includes
> ### 1 input layer with 74 neurons
> ### 1 hidden layer with 74 neurons and with "relu" activation function
> ### 1 output layer with a single neuron with "sigmoid" activation function for binary classification output of "credit_default"

```python
from tensorflow.keras.models import Model
from tensorflow.keras.models import load_model

# Loading the Pre-Trained Target Model 2 (TM2) for Classification Evaluation of "credit_default"

tm2_model_path = "/Users/vidhyalakshmiparthasarathy/.CMVolumes/Google-Drive-pbvidhya/QMUL_MSc_Project/credit_risk_scoring_project/models/target_model/tm_model_2_clas_fe_mod_retrain_2_600_final/tm_model_2_clas_fe_mod_retrain_2_600_final.h5"

target_model_2_pre_trained = load_model(tm2_model_path)

print("Target Model 2 Loaded Successfully")
```

```python
# Verifying the Summary of the Model

target_model_2_pre_trained.summary()
```

## Target Model 2 (TM2) Scenario-2 - Validation Evaluation - Classification - "credit_default"

```python
# Evaluating the Target "credit_default" on the Unseen Test Data

tm2_y_pred = target_model_2_pre_trained.predict(tm2_test_X, batch_size=1)

print("Execution Completed")
```

```python
# Verifying the Count of Evaluation Outcomes on the Unseen Test Data

print("Total Output Actuals on the Source Model Validation Data: ", tm2_test_y.shape[0])
print("Total Output Predictions on the Source Model Validation Data: ", len(tm2_y_pred))
```

```python
# Comparing the Actual and Prediction Results for the first 10 Validation Data Instances

print("Target Model 2 Validation Actuals")
print(tm2_test_y[:10])

print("Target Model 2 Validation Predictions")
print(tm2_y_pred[:10])
```

## Target Model 2 (TM2) - Actuals versus Predictions Comparison

### Converting the Probability Predictions of the "credit_default" into a Binary Value Representation of 0 and 1

```python
# Converting probability predictions numpy array into a dataframe

tm2_y_pred_df = pd.DataFrame(tm2_y_pred, columns=['Credit_Default_Probability_Prediction'])
tm2_y_pred_df.head()
```

```python
tm2_y_pred_df['Credit_Default_Class_Prediction'] = np.where( tm2_y_pred_df['Credit_Default_Probability_Prediction'] >= 0.5, 1, 0)

print("Missing Values in the New Column 'Credit_Default_Class_Prediction':", tm2_y_pred_df['Credit_Default_Class_Prediction'].isna().sum())

print("\n")

# Displaying a sample of 10 random actuals and its respective predictions
tm2_y_pred_df.sample(10)
```

```python
# Verifying the Size and the Missing Values (if any) Count of the Score Status Actual (Ground-Truth) Dataframe and the Predictions Dataframe

print(tm2_test_y.shape)
print(tm2_y_pred_df.shape)

print("\n")

print(tm2_test_y.isna().sum())
print(tm2_y_pred_df.isna().sum())
```

```python
# Concatenating the Credit Default Actual (Ground-Truth) Dataframe and the Predictions Dataframe

frames = [tm2_test_y, tm2_y_pred_df]

tm2_results = pd.concat(frames, axis=1)

# Renaming the "credit_default" Column Name
tm2_results.rename(columns = {'credit_default':'Credit_Default_Class_Actual'}, inplace = True)

# Displaying a sample of 10 random results
tm2_results.sample(10)
```

## Target Model 2 (TM2) - Exporting a Copy of the "credit_default" Output Predictions

```python
# Exporting a copy of Credit Default Actuals and Predictions DataFrame as a csv file

tm2_results.to_csv("/Users/vidhyalakshmiparthasarathy/.CMVolumes/Google-Drive-pbvidhya/QMUL_MSc_Project/credit_risk_scoring_project/output/tm2_test_results/tm2_test_results_optimised_model.csv")

print("Data Export Completed")
```

## Target Model 2 (TM2) - Metrics Report

```python
# Target Model 2 (TM2) Scenario-2 - Classificatio Report 

from sklearn.metrics import confusion_matrix, classification_report

tm2_classification_report = classification_report(tm2_results['Credit_Default_Class_Actual'], tm2_results['Credit_Default_Class_Prediction'])

print(tm2_classification_report)
```

```python
# Target Model 2 (TM2) Scenario-2 - Confusion Matrix

import tensorflow as tf
import seaborn as sn

tm2_cm = tf.math.confusion_matrix(labels=tm2_results['Credit_Default_Class_Actual'], predictions=tm2_results['Credit_Default_Class_Prediction'])

plt.figure(figsize = (8, 6))
sn.heatmap(tm2_cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# Target Model 2 (TM2) Scenario-2 - ROC, AUC

from sklearn.metrics import roc_auc_score

tm2_auc = roc_auc_score(tm2_results['Credit_Default_Class_Actual'], tm2_results['Credit_Default_Class_Prediction'])

print("Source Model ROC AUC: ", tm2_auc)
```
