# **5. Importing the Base Model (Source Model)**

```python
from tensorflow.keras.models import Model
from tensorflow.keras.models import load_model

# Loading an Existing Trained Model

source_model_path = "/content/drive/MyDrive/QMUL_MSc_Project/credit_risk_scoring_project/models/source_model/1_sm_adam_relu_sigmoid_4_0.001_inputsize/1_sm_adam_relu_sigmoid_4_0.001_inputsize.h5"

source_model = load_model(source_model_path)

print("Pre-Trained Source Model Loaded Successfully")
```

```python
# Verifying the Summary of the Existing Pre-Trained Source Model

source_model.summary()
```

# **6. Transfer Learning of the Source Model into the Target Model : Configurations**

```python
# Freezing the Source Model to avoid updating the weights during the Target Model Training

source_model.trainable = False

print("Execution Completed")
```

```python
# Retaining only the First Hidden Layer, Removed 2nd, 3rd, 4th Hidden Layers and Removed Output Layer from the Source Model and Creating a Base Model

from keras.models import Model

base_model = Model(inputs=source_model.input, outputs=source_model.layers[0].output)

print("Execution Completed")                                    
```

```python
# Verifying the Summary of the Base Model after removing the Output (Last) Layer to use the Source Model as Feature Extractor

base_model.summary()
```

**As we can see; output (last) layer is removed from the base model for transfer learning feature extraction purposes.**

# **7. Transfer Learning of the Source Model into the Target Model 1 : Feature Extraction - Model Definition**

```python
# Defining the Target Model from the Pre-Trained Source Model

'''
inputs = base_model
output_reg = keras.layers.Dense(1, activation='linear')
output_clas = keras.layers.Dense(1, activation='sigmoid')

outputs = [output_reg, output_clas]

target_model = tf.keras.Model(inputs=inputs, outputs=output_reg)
'''

target_model_1 = tf.keras.Sequential([
                                    base_model,
                                    #keras.layers.Dense(1, activation='linear')
                                    #keras.layers.Dense(1)
                                    keras.layers.Dense(1, activation='linear')
])

# Compiling the Model

base_lr_1 = 0.001
tm1_lr = 0.025

optim = tf.keras.optimizers.Adam(learning_rate=base_lr_1)
#loss = tf.keras.losses.MeanSquaredError()
loss = 'mean_absolute_error'
metrics = 'mean_squared_error'

target_model_1.compile(
    optimizer=optim,
    loss=loss,
    metrics=[metrics])

print("Execution Completed")                                   
```

```python
# Verifying the Summary of the Target Model (with Pre-Trained Source Model as Feature Extractor)

target_model_1.summary()
```

# **8. Target Model Training - With Source Model as Feature Extractor and Customised**

```python
# Target Model Training - With Source Model as Feature Extractor

# Training Configurations

input_data_size = tm_train_X.shape[0]

epochs_1 = 5000
batch_size_1 = input_data_size

# Tensorboard Configuration

#log_dir = "/content/drive/MyDrive/QMUL_MSc_Project/credit_risk_scoring_project/logs/tm_1_tune_2_retrain_1/"

tb_callback_tm_1 = tf.keras.callbacks.TensorBoard(log_dir="/content/drive/MyDrive/QMUL_MSc_Project/credit_risk_scoring_project/logs/tm_1_tune_2_retrain_1/", 
                                                 histogram_freq=1)


print("Model Training Started.....")

history = target_model_1.fit(tm_train_X, tm_train_y_reg,
                             validation_data=(tm_validation_X, tm_validation_y_reg),
                             epochs=epochs_1,
                             batch_size=batch_size_1, 
                             callbacks=[tb_callback_tm_1])

print("Model Training Completed.....")
```

# **14. Target Model 1 - Accuracy and Error Performance Metrics**

#### Adding a new column "Credit_Score_Deviation_Status" to track whether the deviation between the actuals and predicted Credit Scores are less than or equal to 10 or not.
#### If the deviation is less than or equal to 10; then mark new column "Credit_Score_Deviation_Status" as "Accept".
### If the deviation is greater than 10; then mark new column "Score_Deviation_Status" as "Review".
### If the deviation is 0; then mark new column "Score_Deviation_Status" as "Perfect".

#### **PLEASE NOTE:** The Deviation of 10 is a custom number selected; in order to verify the number of predictions that are closer to the actuals.
#### This is a custom trade-off number which we are using to derive at a agreed model with acceptable accuracy.
#### As input features related to Non-FICO factor named "Lenders Internal" Factor was not in scope of the Actual FICO Scores and this proposed custom credit scoring model includes "Lenders Internal" Factor; hence it is a sensible decision to have a trade-off in the deviation of the predicted scores from that of the actual scores in order to decide on a good credit scoring model.

```python
# Adding a new column "Credit_Score_Deviation_Status" to track whether the deviation between the actuals and predicted Credit Scores are 
# less than or equal to 10 or not.

# If the deviation is less than or equal to 10; then mark new column "Credit_Score_Deviation_Status" as "Accept".
# If the deviation is greater than 10; then mark new column "Score_Deviation_Status" as "Review".
# If the deviation is 0; then mark new column "Score_Deviation_Status" as "Perfect".

# PLEASE NOTE: The Deviation of 10 is a custom number selected; in order to verify the number of predictions that are closer to the actuals.
# This is a custom trade-off number which we are using to derive and select an agreed best-fit model with an acceptable error/deviation.

# As input features related to Non-FICO factor named "Lenders Internal" Factor was not in scope of the Actual FICO Scores 
# and this proposed custom credit scoring model includes "Lenders Internal" Factor; hence it is a sensible decision to have a trade-off in the 
# deviation of the predicted scores from that of the actual scores in order to decide on a good credit scoring model.


tm1_results['Credit_Score_Deviation_Status'] = np.where(tm1_results['Credit_Score_Deviation'] <= 49 , "Accept", "Review")
tm1_results['Credit_Score_Deviation_Status'] = np.where(tm1_results['Credit_Score_Deviation'] == 0, "Perfect", tm1_results['Credit_Score_Deviation_Status'])

# Displaying the first 10 data instances of the data frame
tm1_results.head(10)
```

```python
# Extracting the Total Count of "Accept" and "Review" Credit Score from the "Credit_Score_Deviation_Status" Column

tm1_deviation_status_count = tm1_results['Credit_Score_Deviation_Status'].value_counts()
tm1_deviation_status_count
```

```python
# Calculating the Percentage of Accuracy and Error based on the Acceptable Custom Deviation Value set as an accepted trade-off in terms of the 
# deviations of the predicted scores from that of the actual scores.

tm1_accuracy_percentage = (tm1_deviation_status_count['Accept'] + tm1_deviation_status_count['Perfect']) / tm1_results.shape[0] * 100
print("Target Model 1 - Accuracy in Percentage: ", tm1_accuracy_percentage)

tm1_error_percentage = tm1_deviation_status_count['Review'] / tm1_results.shape[0] * 100
print("Target Model 1 - Error in Percentage: ", tm1_error_percentage)
```
