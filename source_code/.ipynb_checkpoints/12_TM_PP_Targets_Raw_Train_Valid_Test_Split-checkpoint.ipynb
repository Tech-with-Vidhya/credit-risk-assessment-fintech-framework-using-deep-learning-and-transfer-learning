## Splitting the Target Model (TM) Raw Data in to 80% Train Data, 16% Validation Data and 4% Test Data

> ### 1. Splitting the Target Model (TM) Raw Data into 80% Train Data and 20% Remaining Data as per standard data split
> ### 2. Splitting the Target Model (TM) Remaining Data into 80% Validation Data and 20% Test Data as per standard data split

```python
# Shuffling the dataframe randomly based on the data instances (rows)

target_model_pp_targets_raw_data_shuffled = target_model_pp_targets_raw_data.sample(frac=1, random_state=5, axis=0)

print("Execution Completed")
```

```python
# Target Model (TM) Raw Training Data

# 80% Target Model (TM) Raw Training Data

target_model_raw_train = target_model_pp_targets_raw_data_shuffled.sample(frac=0.8, random_state=8, axis=0)
print("Total Target Model (TM) Raw Data Instances: ", len(target_model_pp_targets_raw_data_shuffled))
print("Target Model (TM) Raw Training Data Instances: ", len(target_model_raw_train))     
    
print("Execution Completed")
```

```python
# Target Model (TM) Raw Remaining Data

# 20% Target Model (TM) Raw Remaining Data

target_model_raw_remaining = target_model_pp_targets_raw_data_shuffled.drop(target_model_raw_train.index)
print("Total Target Model (TM) Raw Data Instances: ", len(target_model_pp_targets_raw_data_shuffled))
print("Target Model (TM) Raw Remaining Data Instances: ", len(target_model_raw_remaining))

print("Execution Completed")
```

```python
# Target Model (TM) Raw Validation Data

# 80% of Target Model (TM) Raw Remaining Data, i.e.; 16% of the Overall Target Model (TM) Raw Data

target_model_raw_validation = target_model_raw_remaining.sample(frac=0.8, random_state=12, axis=0)
print("Total Target Model (TM) Raw Remaining Data Instances: ", len(target_model_raw_remaining))
print("Target Model (TM) Raw Validation Data Instances: ", len(target_model_raw_validation))     
    
print("Execution Completed")
```

```python
# Target Model (TM) Raw Test Data

# 20% of Target Model (TM) Raw Remaining Data, i.e.; 4% of the Overall Target Model (TM) Raw Data

target_model_raw_test = target_model_raw_remaining.drop(target_model_raw_validation.index)
print("Total Target Model (TM) Raw Remaining Data Instances: ", len(target_model_raw_remaining))
print("Target Model (TM) Raw Test Data Instances: ", len(target_model_raw_test))

print("Execution Completed")
```

```python
# Verifying the size of the Training, Validation and Test Data Splits of the Target Model (TM) Dataset

print("Target Model (TM) Total Raw Data Size: ", target_model_pp_targets_raw_data.shape)


print("\nTarget Model Training Raw Data Size: ", target_model_raw_train.shape)
print("Target Model Validation Raw Data Size: ", target_model_raw_validation.shape)
print("Target Model Test Raw Data Size: ", target_model_raw_test.shape)
```

```python
# Reset the index of the Raw Training Data of the Target Model (TM) Dataset

target_model_raw_train.reset_index(drop=True, inplace=True)

print("Execution Completed")
```

```python
# Reset the index of the Raw Validation Data of the Target Model (TM) Dataset

target_model_raw_validation.reset_index(drop=True, inplace=True)

print("Execution Completed")
```

```python
# Reset the index of the Raw Test Data of the Target Model (TM) Dataset

target_model_raw_test.reset_index(drop=True, inplace=True)

print("Execution Completed")
```
