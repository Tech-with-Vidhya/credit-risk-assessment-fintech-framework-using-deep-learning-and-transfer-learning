## Weight of Evidence (WoE) : Pre-processing all the Numeric Continuous Data Type Input Variables to Categorical Data Type Variables

### Weight of Evidence (WoE) Technique helps to transform any continuous data type variables in to categorical data type variables. This is achieved by splitting the continuous variables values into one or many bins (as categories) and further based on the individual WoE values for each of the bins; new categorical variables will be created.

### Thumb Rules While Applying Weight of Evidence (WoE) Technique:
> ### 1. Each bin (category) must have atleast 5% of the total observations. 
> ### However from credit scoring domain perspective; having a value as 0 or having a very higher value have a tremendous impact on the scores by either increasing or decreasing the overall score on a case by case basis. Hence by considering the domain significance with respect to credit scoring; it is obsolutely essential to have bins like "observations having 0 value" and "observations having maximum value" to be included as valid bins/categories even if the total number of observations in such a category is less than 5%. 
> ### Hence credit scoring domain needs/requisites takes precedence when compared with the inclusion of this rule; and it's been factored accordingly.
> ### 2. Bins having unique WoE values must be kept as seperate categorical variables and they should not be combined together.
> ### 3. Bins having same or closer WoE values can be combined into one single bin; provided if such bins don't have significant difference in the number of observations (data instances) between them. 
> ### 4. WoE values and curve must be monotonic; i.e.; either increasing or decreasing with the bins (categories).

## Custom Function to Calculate the Weight of Evidence (WoE) for Continuous Input Variables

```python
# Custom Function to Calculate WoE for Continuous Input Variables

def sm_woe_continous(df, new_cat_feature_name, df_score_status):
    df = pd.concat([df[new_cat_feature_name], df_score_status], axis=1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                   df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good+']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good+'] = df['prop_good+'] * df['n_obs']
    df['n_good-'] = (1 - df['prop_good+']) * df['n_obs']
    df['prop_n_good+'] = df['n_good+'] / df['n_good+'].sum()
    df['prop_n_good-'] = df['n_good-'] / df['n_good-'].sum()
    df['WoE'] = np.log(df['prop_n_good+'] / df['prop_n_good-'])
#     df = df.sort_values(['WoE'])
#     df = df.reset_index(drop = True)
    df['diff_prop_good+'] = df['prop_good+'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good+'] - df['prop_n_good-']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df

print("Execution Completed")
```

## Custom Function to Plot the Weight of Evidence (WoE) Values for Continuous Input Variables

```python
# Custom Function to Plot the WoE Values

# Seaborn Style Plt
sns.set()

# WoE Plot Custom Function Definition
def sm_plot_by_woe(df_WoE, rotation_of_x_axis_labels = 0):
    x = np.array(df_WoE.iloc[:, 0].apply(str))
    y = df_WoE['WoE']
    plt.figure(figsize=(18, 6))
    plt.plot(x, y, marker = 'o', linestyle = '--', color = 'k')
    plt.xlabel(df_WoE.columns[0])
    plt.ylabel('Weight of Evidence')
    plt.title(str('Weight of Evidence by ' + df_WoE.columns[0]))
    plt.xticks(rotation = rotation_of_x_axis_labels)
    
print("Execution Completed")
```
