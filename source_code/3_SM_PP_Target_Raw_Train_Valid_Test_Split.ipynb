## Splitting the Source Model (SM) Raw Data in to 80% Train Data, 16% Validation Data and 4% Test Data

> ### 1. Splitting the Source Model (SM) Raw Data into 80% Train Data and 20% Remaining Data as per standard data split
> ### 2. Splitting the Source Model (SM) Remaining Data into 80% Validation Data and 20% Test Data as per standard data split

```python
# Shuffling the dataframe randomly based on the data instances (rows)

source_model_pp_target_raw_data_shuffled = source_model_pp_target_raw_data.sample(frac=1, random_state=5, axis=0)

print("Execution Completed")
```

```python
# Source Model (SM) Raw Training Data

# 80% Source Model (SM) Raw Training Data

source_model_raw_train = source_model_pp_target_raw_data_shuffled.sample(frac=0.8, random_state=8, axis=0)
print("Total Source Model (SM) Raw Data Instances: ", len(source_model_pp_target_raw_data_shuffled))
print("Source Model (SM) Raw Training Data Instances: ", len(source_model_raw_train))     
    
print("Execution Completed")
```

```python
# Source Model (SM) Raw Remaining Data

# 20% Source Model (SM) Raw Remaining Data

source_model_raw_remaining = source_model_pp_target_raw_data_shuffled.drop(source_model_raw_train.index)
print("Total Source Model (SM) Raw Data Instances: ", len(source_model_pp_target_raw_data_shuffled))
print("Source Model (SM) Raw Remaining Data Instances: ", len(source_model_raw_remaining))

print("Execution Completed")
```

```python
# Source Model (SM) Raw Validation Data

# 80% of Source Model (SM) Raw Remaining Data, i.e.; 16% of the Overall Source Model (SM) Raw Data

source_model_raw_validation = source_model_raw_remaining.sample(frac=0.8, random_state=12, axis=0)
print("Total Source Model (SM) Raw Remaining Data Instances: ", len(source_model_raw_remaining))
print("Source Model (SM) Raw Validation Data Instances: ", len(source_model_raw_validation))     
    
print("Execution Completed")
```

```python
# Source Model (SM) Raw Test Data

# 20% of Source Model (SM) Raw Remaining Data, i.e.; 4% of the Overall Source Model (SM) Raw Data

source_model_raw_test = source_model_raw_remaining.drop(source_model_raw_validation.index)
print("Total Source Model (SM) Raw Remaining Data Instances: ", len(source_model_raw_remaining))
print("Source Model (SM) Raw Test Data Instances: ", len(source_model_raw_test))

print("Execution Completed")
```

```python
# Verifying the size of the Training, Validation and Test Data Splits of the Source Model (SM) Dataset

print("Source Model (SM) Total Raw Data Size: ", source_model_pp_target_raw_data.shape)


print("\nSource Model Training Raw Data Size: ", source_model_raw_train.shape)
print("Source Model Validation Raw Data Size: ", source_model_raw_validation.shape)
print("Source Model Test Raw Data Size: ", source_model_raw_test.shape)
```

```python
# Reset the index of the Raw Training Data of the Source Model (SM) Dataset

source_model_raw_train.reset_index(drop=True, inplace=True)

print("Execution Completed")
```

```python
# Reset the index of the Raw Validation Data of the Source Model (SM) Dataset

source_model_raw_validation.reset_index(drop=True, inplace=True)

print("Execution Completed")
```

```python
# Reset the index of the Raw Test Data of the Source Model (SM) Dataset

source_model_raw_test.reset_index(drop=True, inplace=True)

print("Execution Completed")
```
